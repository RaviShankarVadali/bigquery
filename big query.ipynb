```python
res_type = gbq.read_gbq("SELECT RestaurantType , count(RestaurantType) as COUNT_OF FROM `plucky-agency-421911.DEMO.ZOMATO DATA` GROUP BY RestaurantType")
```

```python
res_type
```

# data analysis  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20,20))
plt.pie(x = res_type['COUNT_OF'] , labels=res_type['RestaurantType'] , autopct='%f0.02' )
plt.show()
```

```python
res_type = res_type.drop_duplicates()
```

```python
top_10_type  = res_type.sort_values(by=['COUNT_OF'] , ascending=False).head(10)
```

```python
plt.figure(figsize=(20,20))
plt.pie(x = top_10_type['COUNT_OF'] , labels=top_10_type['RestaurantType'] , autopct='%f0.02' )
plt.show()
```

```python
type_cost = gbq.read_gbq("SELECT RestaurantType , AVG(AverageCost) as AVG_COST FROM `plucky-agency-421911.DEMO.ZOMATO DATA` GROUP BY RestaurantType")
```

```python
type_cost.sort_values(by = ['AVG_COST'] , ascending=False , inplace=True)
```

```python
top_costly_type = type_cost.head(10)
top_costly_type
```

```python
sns.barplot(data = top_costly_type , x = 'RestaurantType' , y = 'AVG_COST')
plt.xticks(rotation = 90)
plt.show()
```

```python
top_notcostly_type = type_cost.tail(10)
top_notcostly_type
```

```python
sns.barplot(data = top_notcostly_type , x = 'RestaurantType' , y = 'AVG_COST')
plt.xticks(rotation = 90)
plt.show()
```

```python
area_wise_cost = gbq.read_gbq("SELECT Area ,  RestaurantType , AVG(AverageCost) as AVG_COST FROM `plucky-agency-421911.DEMO.ZOMATO DATA` GROUP BY RestaurantType , Area")
```

```python
area_wise_cost.dropna(inplace=True)
area_wise_cost.sort_values(by = 'AVG_COST' , ascending=False , inplace=True)
```

```python
area_wise_cost
```

```python
costly_areas = area_wise_cost.head(10)
costly_areas
```

```python
notcostly_areas = area_wise_cost.tail(10)
```

```python
plt.figure(figsize=(10,5) )
sns.barplot(data = costly_areas , x = 'Area' , y = 'AVG_COST' , hue='RestaurantType' )
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
ratings_with_res = gbq.read_gbq("SELECT RestaurantName,  area , RestaurantType , AVG(Rating) as AVG_rating FROM `plucky-agency-421911.DEMO.ZOMATO DATA`GROUP BY RestaurantType , Rating , area ,RestaurantName  order by rating DESC ")
```

```python
ratings_with_res.dropna(inplace=True)
```

```python
ratings_with_res.drop(index=2 , axis=0 , inplace=True)
```

```python
high_ratings = ratings_with_res.head(10)
high_ratings
```

```python
low_ratings = ratings_with_res.tail(10)
low_ratings
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data = high_ratings , x = 'area' , y = 'AVG_rating' , hue='RestaurantType' )
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data = low_ratings , x = 'area' , y = 'AVG_rating' , hue='RestaurantType' )
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```
